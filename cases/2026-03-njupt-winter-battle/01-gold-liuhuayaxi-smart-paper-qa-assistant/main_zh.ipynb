# 基于提示词工程与超长上下文处理的文献问答与对比分析系统

本 Notebook 用于在 aup-learning-cloud Basic GPU Environment 中启动论文问答与对比分析界面。

首次运行时如果 `config/app_config.json` 不存在，系统会自动生成默认配置文件；补充好模型与接口参数后即可导入论文并开始使用。


```python
from pathlib import Path
import importlib
import importlib.util
import site
import subprocess
import sys

ROOT = Path.cwd()
if not (ROOT / 'src').exists() and (ROOT.parent / 'src').exists():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

REQUIRED_MODULES = {
    'ipywidgets': 'ipywidgets',
    'langchain': 'langchain',
    'langchain_openai': 'langchain-openai',
    'langchain_community': 'langchain-community',
    'langchain_chroma': 'langchain-chroma',
    'openai': 'openai',
    'chromadb': 'chromadb',
    'pypdf': 'pypdf',
    'docx': 'python-docx',
    'markdown': 'markdown',
    'tiktoken': 'tiktoken',
    'pydantic': 'pydantic',
    'tenacity': 'tenacity',
}

missing_packages = sorted(
    {
        package_name
        for module_name, package_name in REQUIRED_MODULES.items()
        if importlib.util.find_spec(module_name) is None
    }
)
if missing_packages:
    print('检测到缺少依赖，正在当前用户环境自动安装：')
    print(', '.join(missing_packages))
    subprocess.check_call(
        [sys.executable, '-m', 'pip', 'install', '--user', '-r', str(ROOT / 'requirements.txt')]
    )
    user_site = site.getusersitepackages()
    if user_site not in sys.path:
        sys.path.append(user_site)
    importlib.invalidate_caches()
    print('依赖安装完成。')

from IPython.display import display

from src.config import AppConfig
from src.ui_app import build_app

ROOT
config = AppConfig.from_file(ROOT / 'config' / 'app_config.json')
config.ensure_directories()
app = build_app(config)
display(app)
```
