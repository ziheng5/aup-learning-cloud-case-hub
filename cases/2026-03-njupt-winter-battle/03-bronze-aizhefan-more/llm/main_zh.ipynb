## 什么是语言模型？

语言模型是一种人工智能，能够理解和生成人类语言。它通过学习大量文本数据来掌握单词之间的模式和关系。

我们的模型专为日常对话设计，拥有30亿参数，这意味着它从海量文本数据中学习。

## 加载模型

现在让我们加载预训练模型。我们将使用 `inference` 目录中的 `generator.py` 模块。


```python
import sys
import os

# 将项目根目录添加到Python路径
sys.path.append(os.path.abspath('.'))

from inference.generator import Generator

# 初始化生成器
# 注意：您需要将此替换为您训练模型的路径
generator = Generator(model_path='./checkpoints/final')
```

## 生成文本

现在让我们使用模型生成一些文本。我们将提供一个提示，模型将继续生成文本。


```python
# 从提示生成文本
prompt = '你好，今天怎么样？'

# 生成响应
response = generator.generate(
    prompt=prompt,
    max_new_tokens=100,  # 生成的最大token数
    temperature=0.7,     # 控制随机性（越低越确定）
    top_p=0.95           # 控制多样性（越低越集中）
)

print(f"提示: {prompt}")
print(f"响应: {response}")
```

## 交互式对话

让我们创建一个简单的交互式对话系统，您可以与模型聊天。


```python
# 简单的交互式对话
print("欢迎使用聊天机器人！输入'exit'结束对话。")

while True:
    user_input = input("你: ")
    if user_input.lower() == 'exit':
        print("再见！")
        break
    
    # 生成响应
    response = generator.generate(
        prompt=user_input,
        max_new_tokens=150,
        temperature=0.7
    )
    
    print(f"机器人: {response}")
```

## 理解模型

### 关键组件

1. **Transformer架构**: 模型使用解码器-only Transformer架构，类似于GPT模型。
2. **注意力机制**: 帮助模型在生成输出时关注输入的相关部分。
3. **FlashAttention**: 注意力的优化实现，节省内存。
4. **RoPE位置编码**: 帮助模型理解序列中单词的顺序。
5. **SwiGLU激活**: 一种激活函数，帮助模型学习复杂模式。


### 模型大小

- **3B参数**: 这意味着模型有约30亿可学习权重。
- **32层**: 模型有32层Transformer块。
- **24个注意力头**: 允许模型关注输入的不同方面。
- **4096最大序列长度**: 模型可以处理单个输入中最多4096个token。
