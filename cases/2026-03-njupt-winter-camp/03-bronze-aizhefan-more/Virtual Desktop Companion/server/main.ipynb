## Overview

The AI Desktop Pet server is built with:
- **FastAPI**: Modern Python web framework
- **Ollama**: Local LLM integration
- **ChromaDB**: Vector database for memory
- **Redis**: Optional caching layer

```python
import requests
import json

# Configuration
SERVER_URL = 'http://localhost:8000'

print(f"Server URL: {SERVER_URL}")
```

## API Testing

### Health Check

First, let's check if the server is running.

```python
# Check server health
try:
    response = requests.get(f"{SERVER_URL}/api/v1/health")
    print(f"Server Status: {response.json()['status']}")
    print("Server is running!")
except requests.exceptions.ConnectionError:
    print("Server is not running. Please start it first.")
```

### Creating a Session

```python
# Create a new session
response = requests.post(f"{SERVER_URL}/api/v1/session")
session_id = response.json()['session_id']

print(f"Session ID: {session_id}")
```

### Chat with AI

```python
# Send a message to AI
message = "Hello! What can you do?"

response = requests.post(
    f"{SERVER_URL}/api/v1/chat",
    json={
        "session_id": session_id,
        "message": message,
        "stream": False
    }
)

data = response.json()
print(f"AI Response: {data.get('text', 'No response')}")
print(f"\nEmotion: {data.get('emotion', 'unknown')}")
print(f"Action: {data.get('action', 'idle')}")
print(f"\nEmotion Vector: {data.get('emotion_vector', {})}")
```

## Memory System

The server maintains three types of memory:
- **Working Memory**: Recent conversation history
- **Long-term Memory**: Important information and experiences
- **Deep Memory**: User profile and core memories

```python
# Send multiple messages to build memory
messages = [
    "My name is Alice",
    "I like programming",
    "What's my name?",
    "What do you know about me?"
]

for msg in messages:
    response = requests.post(
        f"{SERVER_URL}/api/v1/chat",
        json={
            "session_id": session_id,
            "message": msg,
            "stream": False
        }
    )
    data = response.json()
    print(f"\nUser: {msg}")
    print(f"AI: {data.get('text', 'No response')}")
```

## RAG System

The server uses Retrieval-Augmented Generation (RAG) to provide expert knowledge about WSL2.

```python
# Ask a WSL2-related question
wsl2_question = "How do I install a new Linux distribution in WSL2?"

response = requests.post(
    f"{SERVER_URL}/api/v1/chat",
    json={
        "session_id": session_id,
        "message": wsl2_question,
        "stream": False
    }
)

data = response.json()
print(f"Question: {wsl2_question}")
print(f"\nAI Response: {data.get('text', 'No response')}")
print(f"\nRAG Context Used: {data.get('rag_used', False)}")
```

## Emotion Analysis

The server analyzes emotions and maps them to Live2D parameters.

```python
# Test different emotions
test_messages = [
    "I'm so happy today!",
    "I feel sad about the news",
    "That makes me angry",
    "I'm just neutral"
]

for msg in test_messages:
    response = requests.post(
        f"{SERVER_URL}/api/v1/chat",
        json={
            "session_id": session_id,
            "message": msg,
            "stream": False
        }
    )
    data = response.json()
    print(f"\nMessage: {msg}")
    print(f"Detected Emotion: {data.get('emotion', 'unknown')}")
    print(f"Live2D Parameters: {data.get('live2d_params', {})}")
```

## WSL2 Command Mode

The server can generate WSL2 commands based on natural language requests.

```python
# Test WSL2 command generation
wsl2_requests = [
    "List files in the current directory",
    "Check disk usage",
    "Create a new directory called 'test'"
]

for request in wsl2_requests:
    response = requests.post(
        f"{SERVER_URL}/api/v1/chat",
        json={
            "session_id": session_id,
            "message": request,
            "mode": "wsl2",
            "stream": False
        }
    )
    data = response.json()
    print(f"\nRequest: {request}")
    print(f"Generated Command: {data.get('wsl2_command', 'No command')}")
    print(f"Explanation: {data.get('text', 'No explanation')}")
```
