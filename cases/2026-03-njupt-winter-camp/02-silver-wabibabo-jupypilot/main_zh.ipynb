```python
import pathlib
from jupypilot.config import Config, OllamaConfig, ToolsConfig
from jupypilot.orchestrator.orchestrator import Orchestrator
from jupypilot.ui.app import JupyPilotApp

# 使用绝对路径，避免浏览器环境下 kernel cwd 与 notebook 所在目录不一致导致 open_file 找不到文件
_root = pathlib.Path(".").resolve()
config = Config(
    repo_path=str(_root),
    artifacts_dir=".jupypilot",
    ollama=OllamaConfig(
        endpoint="http://open-webui-ollama.open-webui:11434/api/chat",
        model="qwen3-coder:30b",
    ),
    tools=ToolsConfig(
        allow_write_files=False,
    ),
)
app = JupyPilotApp(Orchestrator(config))
app.render()
```
