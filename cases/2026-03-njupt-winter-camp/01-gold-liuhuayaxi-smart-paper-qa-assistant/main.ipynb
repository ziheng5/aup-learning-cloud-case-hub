# Literature QA and Comparative Analysis System Based on Prompt Engineering and Long-Context Processing

This notebook launches the paper QA and comparative analysis UI in the aup-learning-cloud Basic GPU Environment.

If `config/app_config.json` does not exist, the notebook will create a default config file automatically. After filling in the model and API settings, you can import papers and start using the assistant.


```python
from pathlib import Path
import importlib
import importlib.util
import site
import subprocess
import sys

ROOT = Path.cwd()
if not (ROOT / 'src').exists() and (ROOT.parent / 'src').exists():
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

REQUIRED_MODULES = {
    'ipywidgets': 'ipywidgets',
    'langchain': 'langchain',
    'langchain_openai': 'langchain-openai',
    'langchain_community': 'langchain-community',
    'langchain_chroma': 'langchain-chroma',
    'openai': 'openai',
    'chromadb': 'chromadb',
    'pypdf': 'pypdf',
    'docx': 'python-docx',
    'markdown': 'markdown',
    'tiktoken': 'tiktoken',
    'pydantic': 'pydantic',
    'tenacity': 'tenacity',
}

missing_packages = sorted(
    {
        package_name
        for module_name, package_name in REQUIRED_MODULES.items()
        if importlib.util.find_spec(module_name) is None
    }
)
if missing_packages:
    print('Detected missing dependencies. Installing into the current user environment:')
    print(', '.join(missing_packages))
    subprocess.check_call(
        [sys.executable, '-m', 'pip', 'install', '--user', '-r', str(ROOT / 'requirements.txt')]
    )
    user_site = site.getusersitepackages()
    if user_site not in sys.path:
        sys.path.append(user_site)
    importlib.invalidate_caches()
    print('Dependency installation completed.')

from IPython.display import display

from src.config import AppConfig
from src.ui_app import build_app

ROOT
config = AppConfig.from_file(ROOT / 'config' / 'app_config.json')
config.ensure_directories()
app = build_app(config)
display(app)
```
